```python
import pandas as pd
```

```python
chunk = pd.read_csv('data/Crime/DS0001/da36851-0001.tsv', delimiter='\t', chunksize=1000)
```

```python
df = chunk.get_chunk()
```

```python
df.head()
```

Things to filter on:
- V20111 (location type-1) == 7 (convenience store) or 23 (service/gas station)
- V20112 (location type-2) == 7 (convenience store) or 23 (service/gas station)
- V20113 (location type-3) == 7 (convenience store) or 23 (service/gas station)

- V20061 (UCR OFFENSE CODE - 1) == 233 (Shoplifting) or 234 (Theft From Building) or
- V20062 (UCR OFFENSE CODE - 2) == 233 (Shoplifting) or 234 (Theft From Building) or
- V20063 (UCR OFFENSE CODE - 3) == 233 (Shoplifting) or 234 (Theft From Building)
Things to output:
- INCDATE : date of incdient
- BH007 : city
- BH008 : state abbr
- BH020 : UCR county
- BH054 : FIPS county

```python
type_mask = (df.V20111 == 7) | (df.V20111 == 23) | (df.V20112 == 7) | (df.V20112 == 23) | (df.V20113 == 7) | (df.V20113 == 23)
loc_mask = (df.V20061 == 233) | (df.V20061 == 234) | (df.V20062 == 233) | (df.V20062 == 234) | (df.V20063 == 233) | (df.V20063 == 234)
mask = type_mask & loc_mask
df_sub = df[['INCDATE', 'BH007', 'BH008', 'BH020', 'BH054']]
```

```python
out = pd.DataFrame(columns=['INCDATE', 'BH007', 'BH008', 'BH020', 'BH054'])
```

```python
out = pd.concat([out, df_sub])
```
