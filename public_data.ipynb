#  Food Deserts

```python
import pandas as pd
import numpy as np
pd.set_option('max_columns', 200)
```

```python
fd = pd.read_csv('Food_Desserts/Food_Desserts_2015.csv')
```

```python
fd[ fd.County == 'Suffolk'].head()
```

```python
fd_sum = fd.groupby(['State', 'County']).sum()[['LAPOP1_10','POP2010']]
```

```python
fd_sum.head()
```

```python
fd_sum['FD_ratio'] = fd_sum.LAPOP1_10 / fd_sum.POP2010 
```

```python
fd_sum.head()
```

```python
zips = pd.read_csv('Food_Desserts/us_postal_codes.csv')
```

```python
zips.head()
```

```python
join = zips.join(fd_sum, ['State', 'County'])
```

```python
join[ join.County == 'Autauga']
```

```python
join = join[['Zip Code', 'LAPOP1_10', 'POP2010', 'FD_ratio']]
```

```python
join.to_pickle('Food_Desserts/FD_clean.pkl')
```

```python
unemp = pd.read_csv('Unemployment/Unemployment_2011.csv')
del unemp['# in sample']
unemp.dropna(axis=0, inplace=True)
```

```python
unemp.head()
```

# Crime

```python
chunk = pd.read_csv('data/Crime/DS0001/da36851-0001.tsv', delimiter='\t', chunksize=10000)
out = pd.DataFrame(columns=['INCDATE', 'BH007', 'BH008', 'BH020', 'BH054'])

for df in chunk:
    type_mask = (df.V20111 == 7) | (df.V20111 == 23) | (df.V20112 == 7) | (df.V20112 == 23) | (df.V20113 == 7) | (df.V20113 == 23)
    loc_mask = (df.V20061 == 233) | (df.V20061 == 234) | (df.V20062 == 233) | (df.V20062 == 234) | (df.V20063 == 233) | (df.V20063 == 234)
    mask = type_mask & loc_mask
    
    df_sub = df[['INCDATE', 'BH007', 'BH008', 'BH020', 'BH054']]
    out = pd.concat([out, df_sub])
```

```python
len(out)
```

```python
out.to_pickle('data/Crime/crime_clean.pkl')
```

```python
out['date'] = pd.to_datetime(out['INCDATE'], format='%Y%m%d')
```

```python
out['city'] = out.BH007
out['state'] = out.BH008
```
